## Padding / Border

The method `cv.copyMakeBorder()` is used for adding border around the image(wrapping instead of overlapping). But it is also useful for adding padding.


For `borderType` - Flag defining what kind of border to be added. It can be following [types](https://docs.opencv.org/4.x/d2/de8/group__core__array.html#ga209f2f4869e304c82d07739337eae7c5):
- cv.BORDER_CONSTANT - Adds a constant colored border. The value should be given as next argument.
- cv.BORDER_REFLECT - Border will be mirror reflection of the border elements, like this : fedcba|abcdefgh|hgfedcb
- cv.BORDER_REFLECT_101 or cv.BORDER_DEFAULT - Same as above, but with a slight change, like this : gfedcb|abcdefgh|gfedcba
- cv.BORDER_REPLICATE - Last element is replicated throughout, like this: aaaaaa|abcdefgh|hhhhhhh
- cv.BORDER_WRAP - Can't explain, it will look like this : cdefgh|abcdefgh|abcdefg



You can refer to [here](https://docs.opencv.org/4.x/d3/df2/tutorial_py_basic_ops.html#:~:text=Making%20Borders%20for%20Images%20(Padding)) for more detail.

```python
# Different type of bording
pad_width = 100
replicate = cv.copyMakeBorder(test_img,pad_width,pad_width,pad_width,pad_width,cv.BORDER_REPLICATE)
reflect = cv.copyMakeBorder(test_img,pad_width,pad_width,pad_width,pad_width,cv.BORDER_REFLECT)
reflect101 = cv.copyMakeBorder(test_img,pad_width,pad_width,pad_width,pad_width,cv.BORDER_REFLECT_101)
wrap = cv.copyMakeBorder(test_img,pad_width,pad_width,pad_width,pad_width,cv.BORDER_WRAP)
BLUE = [255,0,0]
constant= cv.copyMakeBorder(test_img,pad_width,pad_width,pad_width,pad_width,cv.BORDER_CONSTANT,value=BLUE)

plt.figure(figsize=(15,10))
plt.subplot(231),plt_show(test_img),plt.title('ORIGINAL')
plt.subplot(232),plt_show(replicate),plt.title('REPLICATE')
plt.subplot(233),plt_show(reflect),plt.title('REFLECT')
plt.subplot(234),plt_show(reflect101),plt.title('REFLECT_101')
plt.subplot(235),plt_show(wrap),plt.title('WRAP')
plt.subplot(236),plt_show(constant),plt.title('CONSTANT')
print ("Original image : ", test_img.shape)
print ("Padded image : ", replicate.shape)
```
